<h2 id="a">¿Qué es? </h2>


Un algoritmo de clustering es una técnica de aprendizaje no supervisado que agrupa un conjunto de objetos de 
tal manera que los objetos dentro de un mismo grupo (o clúster) son más similares entre sí que con aquellos 
de otros grupos. La similitud entre los objetos puede ser determinada en función de diversas métricas, como 
la distancia euclidiana, la correlación, entre otros.

<h2 id="b">¿Cómo funciona un algoritmo de clustering?</h2>

__Inicialización:__ Seleccionar los puntos iniciales que actuarán como los centros de los clústeres.

__Asignación:__ Asignar cada punto de datos al clúster más cercano según una medida de similitud (por ejemplo, distancia euclidiana).

__Actualización:__ Recalcular la posición de los centros de los clústeres basándose en los puntos asignados a cada clúster.

__Iteración:__ Repetir los pasos de asignación y actualización hasta que los centros de los clústeres ya no cambien significativamente 
(convergencia) o se alcance un número máximo de iteraciones.


<h2 id="c">Tipos de algoritmos de clustering</h2>

__K-means:__

Divide el conjunto de datos en 𝐾 clústeres, minimizando la variación interna dentro de cada clúster.
Muy popular por su simplicidad y eficiencia en grandes conjuntos de datos.



__Jerárquico:__

Crea un árbol de clústeres mediante un enfoque aglomerativo (de abajo hacia arriba) o divisivo (de arriba hacia abajo).
No requiere especificar el número de clústeres de antemano.



__DBSCAN (Density-Based Spatial Clustering of Applications with Noise):__

Encuentra clústeres basados en la densidad de puntos en el espacio de datos.
Eficaz para detectar clústeres de formas arbitrarias y manejar ruido en los datos.



__Mean-Shift:__

Desplaza iterativamente los puntos hacia la media de los puntos circundantes en una ventana de radio fijo.
No requiere especificar el número de clústeres de antemano.



__Gaussian Mixture Models (GMM):__

Modela los datos como una mezcla de varias distribuciones gaussianas.
Utiliza el algoritmo EM (Expectation-Maximization) para encontrar los parámetros de las gaussianas.




<h2 id="d">Ejemplificarlo con un diagrama</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Generar datos aleatorios
np.random.seed(0)
X = np.random.rand(100, 2)

# Configuración de KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Predicciones de clústeres
y_kmeans = kmeans.predict(X)

# Centros de los clústeres
centers = kmeans.cluster_centers_

# Crear el diagrama
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Paso 1: Inicialización
ax[0].scatter(X[:, 0], X[:, 1], c='grey', s=50)
ax[0].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='*')
ax[0].set_title('Paso 1: Inicialización de centros')

# Paso 2: Asignación
ax[1].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
ax[1].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='*')
ax[1].set_title('Paso 2: Asignación a clústeres')

# Paso 3: Actualización
# Aquí mostramos los puntos y los centros finales
for i, color in enumerate(['purple', 'orange', 'green']):
    points = X[y_kmeans == i]
    ax[2].scatter(points[:, 0], points[:, 1], s=50, color=color)
    ax[2].scatter(centers[i, 0], centers[i, 1], s=200, color=color, edgecolor='black', marker='*')
ax[2].set_title('Paso 3: Actualización de centros')

plt.show()
```

<h2 id="a">¿Qué es? </h2>

Un algoritmo basado en probabilidad es una técnica de aprendizaje automático que utiliza modelos probabilísticos 
para realizar inferencias y predicciones. Estos algoritmos se basan en la teoría de la probabilidad para modelar 
la incertidumbre en los datos y para hacer suposiciones sobre la estructura subyacente de los datos. En lugar de 
hacer asignaciones deterministas, estos modelos proporcionan distribuciones de probabilidad para sus predicciones.

<h2 id="b">¿Cómo funciona un algoritmo basado en probabilidad?</h2>

__Definición del modelo:__ Se elige un modelo probabilístico que describa la distribución de los datos. Por ejemplo, una distribución gaussiana para modelos de mezcla gaussiana.

__Estimación de parámetros:__ Se utilizan técnicas como la estimación de máxima verosimilitud (MLE) o el algoritmo Expectation-Maximization (EM) para estimar los parámetros del modelo a partir de los datos de entrenamiento.

__Inferencia:__ Se realizan inferencias sobre nuevas observaciones utilizando el modelo probabilístico. Esto puede incluir la asignación de probabilidades a diferentes clases en clasificación o la estimación de valores en regresión.

__Actualización:__ En algunos casos, los modelos se pueden actualizar en línea con nuevas observaciones, ajustando los parámetros de acuerdo con la nueva evidencia.


<h2 id="c">c. Tipos de algoritmos basados en probabilidad
</h2>

__Modelos de Mezcla Gaussiana (GMM):__

Modelan los datos como una combinación de varias distribuciones gaussianas.
Utilizan el algoritmo EM para encontrar los parámetros de cada componente gaussiana.


__Redes Bayesianas:__

Representan relaciones probabilísticas entre un conjunto de variables usando un grafo dirigido acíclico.
Se utilizan para inferencia y toma de decisiones bajo incertidumbre.


__Naive Bayes:__

Un clasificador probabilístico basado en el teorema de Bayes con la suposición de independencia entre características.
Comúnmente utilizado en clasificación de textos y detección de spam.


__Modelos Ocultos de Markov (HMM):__

Modelan sistemas donde se asume que hay un proceso oculto que evoluciona con el tiempo.
Utilizados en reconocimiento de voz, bioinformática y otros campos.
