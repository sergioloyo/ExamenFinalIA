Un algoritmo de clustering es una t칠cnica de aprendizaje no supervisado que agrupa un conjunto de objetos de 
tal manera que los objetos dentro de un mismo grupo (o cl칰ster) son m치s similares entre s칤 que con aquellos 
de otros grupos. La similitud entre los objetos puede ser determinada en funci칩n de diversas m칠tricas, como 
la distancia euclidiana, la correlaci칩n, entre otros.

<h2 id="b">쮺칩mo funciona un algoritmo de clustering?</h2>

__Inicializaci칩n:__ Seleccionar los puntos iniciales que actuar치n como los centros de los cl칰steres.

__Asignaci칩n:__ Asignar cada punto de datos al cl칰ster m치s cercano seg칰n una medida de similitud (por ejemplo, distancia euclidiana).

__Actualizaci칩n:__ Recalcular la posici칩n de los centros de los cl칰steres bas치ndose en los puntos asignados a cada cl칰ster.

__Iteraci칩n:__ Repetir los pasos de asignaci칩n y actualizaci칩n hasta que los centros de los cl칰steres ya no cambien significativamente 
(convergencia) o se alcance un n칰mero m치ximo de iteraciones.


<h2 id="c">Tipos de algoritmos de clustering</h2>

__K-means:__

Divide el conjunto de datos en 洧 cl칰steres, minimizando la variaci칩n interna dentro de cada cl칰ster.
Muy popular por su simplicidad y eficiencia en grandes conjuntos de datos.



__Jer치rquico:__

Crea un 치rbol de cl칰steres mediante un enfoque aglomerativo (de abajo hacia arriba) o divisivo (de arriba hacia abajo).
No requiere especificar el n칰mero de cl칰steres de antemano.



__DBSCAN (Density-Based Spatial Clustering of Applications with Noise):__

Encuentra cl칰steres basados en la densidad de puntos en el espacio de datos.
Eficaz para detectar cl칰steres de formas arbitrarias y manejar ruido en los datos.



__Mean-Shift:__

Desplaza iterativamente los puntos hacia la media de los puntos circundantes en una ventana de radio fijo.
No requiere especificar el n칰mero de cl칰steres de antemano.



__Gaussian Mixture Models (GMM):__

Modela los datos como una mezcla de varias distribuciones gaussianas.
Utiliza el algoritmo EM (Expectation-Maximization) para encontrar los par치metros de las gaussianas.




<h2 id="d">Ejemplificarlo con un diagrama</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Generar datos aleatorios
np.random.seed(0)
X = np.random.rand(100, 2)

# Configuraci칩n de KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# Predicciones de cl칰steres
y_kmeans = kmeans.predict(X)

# Centros de los cl칰steres
centers = kmeans.cluster_centers_

# Crear el diagrama
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Paso 1: Inicializaci칩n
ax[0].scatter(X[:, 0], X[:, 1], c='grey', s=50)
ax[0].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='*')
ax[0].set_title('Paso 1: Inicializaci칩n de centros')

# Paso 2: Asignaci칩n
ax[1].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
ax[1].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5, marker='*')
ax[1].set_title('Paso 2: Asignaci칩n a cl칰steres')

# Paso 3: Actualizaci칩n
# Aqu칤 mostramos los puntos y los centros finales
for i, color in enumerate(['purple', 'orange', 'green']):
    points = X[y_kmeans == i]
    ax[2].scatter(points[:, 0], points[:, 1], s=50, color=color)
    ax[2].scatter(centers[i, 0], centers[i, 1], s=200, color=color, edgecolor='black', marker='*')
ax[2].set_title('Paso 3: Actualizaci칩n de centros')

plt.show()
```
